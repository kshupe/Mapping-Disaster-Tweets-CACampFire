```python
from twitterscraper import query_tweets
import datetime
```

```python
query = "fire OR campfire near: near:39.76,-121.6 within:650mi"
limit_tweets = 100
begin_date = datetime.date(2018, 11, 8)
end_date = datetime.date(2018, 11, 25)
lang = "english"

list_tweets = query_tweets(query,
                           limit=limit_tweets,
                           begindate=begin_date,
                           enddate=end_date)
```

```python
# EACH TWEET IS OF A .TWEET CLASS OBJECT 

for tweet in list_tweets:
    print(type(tweet))
```

```python
df = pd.DataFrame(tweet.__dict__ for tweet in list_tweets)
```
