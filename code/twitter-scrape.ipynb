```python
import pandas as pd
from twitterscraper import query_tweets
import datetime as dt
```

## Twitterscraper Object

**This approach retrieves relevant tweets according to requests passed into a query into Twitterscraper**

Some notes on the scraping tool:
    
1) Tweets can be geo-located by entering the geo-coordinates directly into the query in a `latitude,longitude,radius` format. 
    
-**Note: The only function of location is as input into the `query` to retrieve relevant tweets. User location or tweet location is not available as an output of the scrape.**

2) Twitter users have the option to enable or disable sharing of their location information when using the Twitter platform. As such, searching for tweets based on location criteria is will limit the number of tweets returned by the `query`.

3) For the `query`, a list of commonly used `hashtags` were used to search for users tweeting about relevent subject matter.

**[Reference 1](http://ataspinar.com/2015/11/09/collecting-data-from-twitter/)**
**[Reference 2](https://github.com/taspinar/twitterscraper/issues/109)**
**[Reference 3](https://www.tagsfinder.com/en-us/related/californiafire/)**

```python
# EDIT THESE VARIABLES TO ADJUST THE TWITTER SCRAPE
query = ("fire OR fires OR paradisefire OR paradise OR campfire OR firevictims OR firevictim OR victims OR victim OR calfire OR cafire OR californiafire OR smoke near:39.76,-121.6 within:650mi")

begin_date = dt.date(2018, 11, 8)
end_date = dt.date(2018, 11, 25)

#----------------------------------------------------------
# QUERY 
list_tweets = query_tweets(
    query,
    begindate=begin_date,
    enddate=end_date,
)
```

```python
# CREATE THE DATAFRAME

# REFERENCE:
# https://docs.python.org/3/reference/datamodel.html

raw_df = pd.DataFrame(tweet.__dict__ for tweet in list_tweets)
```

```python
most_recent = raw_df["timestamp"].max()
older_tweet = raw_df["timestamp"].min()

print(f"Raw DF Shape: {raw_df.shape}\n")
print(f"Most Recent Tweet: {most_recent}")
print(f"Oldest Tweet: {older_tweet}\n")
print(f"Col Names & Missing  Values: \n{raw_df.isna().sum()}\n")
print(f"Sample of Text: \n{raw_df.text[:5]}\n")
print(f"Sample: \n{raw_df.text[1]}")
```

```python
#  SAVE CSV  
raw_df.to_csv("../datasets/raw_df.csv")
```
